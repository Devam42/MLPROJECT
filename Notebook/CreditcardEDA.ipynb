### Exploring Data (Visualization)

Exploring the distribution of each feature, especially for features V1 to V28.

Creating box plots to identify potential outliers in the features.

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(df.columns[:-1][:28], 1):
    plt.subplot(7, 4, i)
    sns.boxplot(x='Class', y=column, data=df)
    plt.title(f'Boxplot of {column} by Class')
plt.tight_layout()
```

### Digging the data more to takeout insinghts

### 2. Transaction Amount Statistics

### 3. Time-based Analysis

```python
# Analyzing the distribution of transactions over time
df['Hour'] = df['Time'] // 3600  # Convert time to hours
hourly_stats = df.groupby(['Hour', 'Class']).size().unstack().fillna(0)
print(hourly_stats)
```

```python
# Analyzing the distribution of transactions over time
df['Hour'] = df['Time'] // 3600  # Convert time to hours
hourly_stats = df.groupby(['Hour', 'Class']).size().unstack().fillna(0)

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(hourly_stats[0], label='Normal Transactions', marker='o')
plt.plot(hourly_stats[1], label='Fraud Transactions', marker='o')
plt.title('Distribution of Transactions Over Time')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Transactions')
plt.legend()
plt.grid(True)
```

### 4. Correlation with Target Variable

```python
# Calculating the correlation of features with the target variable (Class)
feature_correlation = df.corrwith(df['Class']).sort_values(ascending=False)
print(feature_correlation)
```

```python
plt.figure(figsize=(10, 6))
feature_correlation.plot(kind='bar', color=['blue' if corr > 0 else 'red' for corr in feature_correlation])
plt.title('Feature Correlation with Class')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.xticks(rotation=45, ha='right')
```

### 5. Outlier Detection

```python
from sklearn.ensemble import IsolationForest

# Identify outliers using Isolation Forest
model = IsolationForest(contamination=0.01)  # Adjust contamination based on your dataset
outliers = model.fit_predict(df.drop(['Time', 'Class'], axis=1))
df['Outlier'] = outliers
outliers_summary = df.groupby('Outlier')['Class'].value_counts()
print(outliers_summary)
```

```python
# Identify outliers using Isolation Forest
model = IsolationForest(contamination=0.01, random_state=42)  # Adjust contamination based on your dataset
outliers = model.fit_predict(df.drop(['Time', 'Class'], axis=1))
df['Outlier'] = outliers

# Plotting
plt.figure(figsize=(8, 6))
df['Outlier'].value_counts().sort_index().plot(kind='bar', color=['blue', 'red'])
plt.title('Outlier Distribution')
plt.xlabel('Outlier')
plt.ylabel('Count')
plt.xticks([0, 1], ['Normal', 'Outlier'], rotation=0)
```

## Conclusion

1. Class Imbalance: The dataset exhibits a significant class imbalance, with a small percentage of transactions labeled as fraud. This imbalance poses challenges for model training, and appropriate techniques like resampling or model adjustments may be needed.
2. Transaction Amounts: Fraudulent transactions tend to have higher transaction amounts compared to normal transactions. Further investigation into the patterns of high-value transactions and their correlation with fraud is recommended.
3. Temporal Patterns: The distribution of transactions over time shows variations, and it's crucial to consider temporal patterns in fraud detection strategies. Further analysis of time-based features and their impact on fraud occurrence could enhance model performance.
4. Feature Correlation: Certain features, such as V4, V11, and V2, show notable correlations with the target variable (fraud). These features may play a crucial role in distinguishing fraudulent transactions and should be prioritized during feature selection for model building.
5. Outlier Detection: Utilizing an Isolation Forest for outlier detection reveals a small percentage of transactions classified as outliers. Investigating these outliers and their corresponding features can provide insights into potential anomalous patterns that may indicate fraudulent behavior.
